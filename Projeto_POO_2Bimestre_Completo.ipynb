## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**Tarefas:**
- Implementar polimorfismo: sobrescrever um m√©todo na subclasse.
- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inser√ß√£o (INSERT), consulta (SELECT) e exclus√£o (DELETE) de registros no banco de dados.

**C√≥digo:**

```python
import sqlite3

class Humor_Aluno_Neo:
    def __init__(self, nome, idade, humor):
        self.nome = nome
        self.idade = idade
        self.humor = humor

    def exibir_dados_Neo(self):
        print(f"{self.nome} | Idade: {self.idade} | Humor: {self.humor}")

class Aluno_Especial_Neo(Humor_Aluno_Neo):
    def __init__(self, nome, idade, humor, atividade_extra):
        super().__init__(nome, idade, humor)
        self.atividade_extra = atividade_extra

    def exibir_dados_Neo(self):
        print(f"{self.nome} (Especial) | Idade: {self.idade} | Humor: {self.humor} | Atividade: {self.atividade_extra}")

con = sqlite3.connect("alunos.db")
cur = con.cursor()

cur.execute("""
CREATE TABLE IF NOT EXISTS alunos (
    nome TEXT,
    idade INTEGER,
    humor TEXT
)
""")

def inserir_aluno(nome, idade, humor):
    cur.execute("INSERT INTO alunos VALUES (?, ?, ?)", (nome, idade, humor))
    con.commit()

def listar_alunos():
    cur.execute("SELECT * FROM alunos")
    for row in cur.fetchall():
        print(row)

def deletar_aluno(nome):
    cur.execute("DELETE FROM alunos WHERE nome = ?", (nome,))
    con.commit()

inserir_aluno("Lucas", 16, "feliz")
inserir_aluno("Ana", 17, "cansada")
print("üìã Lista de Alunos:")
listar_alunos()

print("\n‚ùå Removendo Ana...")
deletar_aluno("Ana")

print("\nüìã Lista Atualizada:")
listar_alunos()

print("\nüé≠ Polimorfismo:")
a1 = Humor_Aluno_Neo("Tiago", 16, "triste")
a2 = Aluno_Especial_Neo("Lara", 17, "feliz", "Teatro")
a1.exibir_dados_Neo()
a2.exibir_dados_Neo()

con.close()
```

```python
# ---------------------------------------------
# PARTE 1 ‚Äì Cria√ß√£o de Classes e Objetos
# ---------------------------------------------

class Humor_Aluno_Neo:
    def __init__(self, nome, idade, humor):
        self.__nome = nome                # Atributo privado (Encapsulamento)
        self.idade = idade
        self.humor = humor

    # ---------------------------------------------
    # PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o
    # ---------------------------------------------

    # Getter e Setter para nome (atributo privado)
    def get_nome(self):
        return self.__nome

    def set_nome(self, novo_nome):
        if novo_nome:
            self.__nome = novo_nome

    # M√©todo com l√≥gica interna
    def verificar_humor(self):
        if self.humor.lower() in ["feliz", "animado"]:
            return "Dia produtivo!"
        elif self.humor.lower() in ["triste", "cansado"]:
            return "Talvez precise descansar..."
        else:
            return "Humor desconhecido."

    def exibir_dados_Neo(self):
        print(f"{self.get_nome()} | Idade: {self.idade} | Humor: {self.humor}")

# ---------------------------------------------
# PARTE 2 ‚Äì Heran√ßa e Subclasses
# ---------------------------------------------

class Aluno_Especial_Neo(Humor_Aluno_Neo):
    def __init__(self, nome, idade, humor, atividade_extra):
        super().__init__(nome, idade, humor)
        self.atividade_extra = atividade_extra

    # PARTE 4 ‚Äì Polimorfismo: sobrescrevendo m√©todo da superclasse
    def exibir_dados_Neo(self):
        print(f"{self.get_nome()} (Especial) | Idade: {self.idade} | Humor: {self.humor} | Atividade: {self.atividade_extra}")

# ---------------------------------------------
# PARTE 4 ‚Äì Integra√ß√£o com Banco de Dados (SQLite)
# ---------------------------------------------
import sqlite3

# Conex√£o com banco (cria arquivo se n√£o existir)
con = sqlite3.connect("alunos.db")
cur = con.cursor()

# Criar tabela
cur.execute("""
CREATE TABLE IF NOT EXISTS alunos (
    nome TEXT,
    idade INTEGER,
    humor TEXT
)
""")

# Fun√ß√£o INSERT
def inserir_aluno(nome, idade, humor):
    cur.execute("INSERT INTO alunos VALUES (?, ?, ?)", (nome, idade, humor))
    con.commit()

# Fun√ß√£o SELECT
def listar_alunos():
    cur.execute("SELECT * FROM alunos")
    for row in cur.fetchall():
        print(row)

# Fun√ß√£o DELETE
def deletar_aluno(nome):
    cur.execute("DELETE FROM alunos WHERE nome = ?", (nome,))
    con.commit()

# ---------------------------------------------
# PARTE 5 ‚Äì Testes Finais
# ---------------------------------------------
print("üîÑ Inserindo alunos no banco:")
inserir_aluno("Lucas", 16, "feliz")
inserir_aluno("Ana", 17, "cansado")

print("\nüìã Alunos no banco:")
listar_alunos()

print("\n‚ùå Deletando Ana do banco...")
deletar_aluno("Ana")

print("\nüìã Alunos ap√≥s exclus√£o:")
listar_alunos()

# Criando objetos e testando POO
print("\nüéì Objetos e Polimorfismo:")
a1 = Humor_Aluno_Neo("Tiago", 16, "triste")
a2 = Aluno_Especial_Neo("Lara", 17, "feliz", "Teatro")

a1.exibir_dados_Neo()
print("‚û°Ô∏è Humor:", a1.verificar_humor())
a2.exibir_dados_Neo()
print("‚û°Ô∏è Humor:", a2.verificar_humor())

# Encerrando conex√£o com banco
con.close()
```
