## PARTE 2 ‚Äì M√©todos e Heran√ßa

**Tarefas:**
- Criar pelo menos 2 m√©todos de a√ß√£o para a classe principal.
- Criar uma subclasse herdando da classe principal.
- Adicionar pelo menos 1 novo atributo e 1 novo m√©todo exclusivo na subclasse.
- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**

```python

class Humor_Aluno_Neo:
    def __init__(self, nome, idade, humor):
        self._nome = nome
        self._idade = idade
        self._humor = humor

    def exibir_dados_Neo(self):
        print(f"Nome: {self._nome}")
        print(f"Idade: {self._idade}")
        print(f"Humor: {self._humor}")
        print("--------------------------")

    # M√©todo de a√ß√£o 1
    def mudar_humor_Neo(self, novo_humor):
        self._humor = novo_humor
        print(f"{self._nome} teve o humor alterado para: {self._humor}")

    # M√©todo de a√ß√£o 2
    def descansar_Neo(self):
        self._humor = "relaxado"
        print(f"{self._nome} descansou e agora est√°: {self._humor}")


# Subclasse herdando da classe principal
class Aluno_Especial_Neo(Humor_Aluno_Neo):
    def __init__(self, nome, idade, humor, atividade_extra):
        super().__init__(nome, idade, humor)  # uso de super()
        self._atividade_extra = atividade_extra  # novo atributo exclusivo

    # Novo m√©todo exclusivo da subclasse
    def mostrar_atividade_Neo(self):
        print(f"{self._nome} participa da atividade extra: {self._atividade_extra}")


# Testando as classes
aluno1 = Humor_Aluno_Neo("Gabriel", 16, "triste")
aluno1.exibir_dados_Neo()
aluno1.descansar_Neo()
aluno1.mudar_humor_Neo("motivado")
print()

aluno2 = Aluno_Especial_Neo("Larissa", 17, "feliz", "Teatro")
aluno2.exibir_dados_Neo()
aluno2.mostrar_atividade_Neo()
```

## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).
- Criar m√©todos `get` e `set` para o atributo privado.
- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**'

```python

class Humor_Aluno_Neo:
    def __init__(self, nome, idade, humor):
        self.__nome = nome           # atributo privado
        self.__idade = idade         # atributo privado
        self.__humor = humor         # atributo privado

    # --------- M√âTODOS GET E SET ---------

    def get_nome_Neo(self):
        return self.__nome

    def get_idade_Neo(self):
        return self.__idade

    def get_humor_Neo(self):
        return self.__humor

    def set_humor_Neo(self, novo_humor):
        self.__humor = novo_humor

    # --------- M√âTODO COM L√ìGICA INTERNA (ABSTRA√á√ÉO) ---------
    def verificar_situacao_Neo(self):
        if self.__humor == "feliz" or self.__humor == "motivado":
            return "Aluno em bom estado emocional. Pode aprender bem!"
        elif self.__humor == "triste" or self.__humor == "cansado":
            return "Aluno pode estar com baixo rendimento. Observar."
        else:
            return "Estado emocional incerto. Acompanhar mais de perto."

    def exibir_dados_Neo(self):
        print("---- DADOS DO ALUNO ----")
        print(f"Nome: {self.__nome}")
        print(f"Idade: {self.__idade}")
        print(f"Humor: {self.__humor}")
        print(f"Situa√ß√£o: {self.verificar_situacao_Neo()}")
        print("-------------------------\n")


# Criando e testando o objeto com verifica√ß√£o
aluno1 = Humor_Aluno_Neo("Eduarda", 16, "cansado")
aluno1.exibir_dados_Neo()

# Mudando o humor usando o set
aluno1.set_humor_Neo("feliz")
aluno1.exibir_dados_Neo()
```

## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**Tarefas:**
- Implementar polimorfismo: sobrescrever um m√©todo na subclasse.
- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inser√ß√£o (INSERT), consulta (SELECT) e exclus√£o (DELETE) de registros no banco de dados.

**C√≥digo:**

```python


import mysql.connector


conexao = mysql.connector.connect(
    host="localhost",
    user="root",               # ou outro nome de usu√°rio
    password="",               # sua senha (pode ser vazia no XAMPP)
    database="banco_neo"       # nome do banco que voc√™ criou no MySQL
)

cursor = conexao.cursor()



class Humor_Aluno_Neo:
    def __init__(self, nome, idade, humor):
        self.nome = nome
        self.idade = idade
        self.humor = humor

    def exibir_dados_Neo(self):
        return f"{self.nome} ({self.idade} anos) - Humor: {self.humor}"

class Aluno_Especial_Neo(Humor_Aluno_Neo):
    def __init__(self, nome, idade, humor, atividade_extra):
        super().__init__(nome, idade, humor)
        self.atividade_extra = atividade_extra

    def exibir_dados_Neo(self):  # polimorfismo
        return f"{self.nome} ({self.idade} anos) - Humor: {self.humor} | Atividade: {self.atividade_extra}"



# Criar tabela (somente 1 vez)
cursor.execute("""
    CREATE TABLE IF NOT EXISTS alunos (
        id INT AUTO_INCREMENT PRIMARY KEY,
        nome VARCHAR(100),
        idade INT,
        humor VARCHAR(50),
        atividade_extra VARCHAR(100)
    )
""")

# Inserir aluno
def inserir_aluno(nome, idade, humor, atividade_extra=None):
    sql = "INSERT INTO alunos (nome, idade, humor, atividade_extra) VALUES (%s, %s, %s, %s)"
    valores = (nome, idade, humor, atividade_extra)
    cursor.execute(sql, valores)
    conexao.commit()
    print(f"‚úÖ Aluno '{nome}' inserido com sucesso no MySQL!")

# Listar alunos
def listar_alunos():
    cursor.execute("SELECT * FROM alunos")
    resultados = cursor.fetchall()
    print("üìã Lista de alunos:")
    for aluno in resultados:
        print(aluno)

# Excluir aluno
def excluir_aluno(id):
    cursor.execute("DELETE FROM alunos WHERE id = %s", (id,))
    conexao.commit()
    print(f"‚ùå Aluno com ID {id} exclu√≠do do banco.")

# ------------------------------
# TESTANDO
# ------------------------------
a1 = Humor_Aluno_Neo("Bruno", 16, "triste")
a2 = Aluno_Especial_Neo("Carla", 17, "feliz", "M√∫sica")

print(a1.exibir_dados_Neo())
print(a2.exibir_dados_Neo())

inserir_aluno(a1.nome, a1.idade, a1.humor)
inserir_aluno(a2.nome, a2.idade, a2.humor, a2.atividade_extra)

listar_alunos()

# excluir_aluno(1)  # Exemplo de exclus√£o

conexao.close()
```

```python
import mysql.connector


conexao = mysql.connector.connect(
    host="localhost",
    user="root",          # Altere se necess√°rio
    password="",          # Senha, se tiver
    database="banco_neo"  # Banco criado no MySQL
)

cursor = conexao.cursor()


cursor.execute("""
CREATE TABLE IF NOT EXISTS alunos (
    id INT AUTO_INCREMENT PRIMARY KEY,
    nome VARCHAR(100),
    idade INT,
    humor VARCHAR(50),
    atividade_extra VARCHAR(100)
)
""")


class Humor_Aluno_Neo:
    def __init__(self, nome, idade, humor):
        # Encapsulamento usando atributos privados
        self.__nome = nome
        self.__idade = idade
        self.__humor = humor

    # GETs e SETs
    def get_nome(self):
        return self.__nome

    def get_idade(self):
        return self.__idade

    def get_humor(self):
        return self.__humor

    def set_humor(self, novo_humor):
        self.__humor = novo_humor

    # M√©todo com l√≥gica interna (Abstra√ß√£o)
    def verificar_situacao(self):
        if self.__humor in ["feliz", "motivado"]:
            return "Aluno pronto para aprender!"
        elif self.__humor in ["triste", "cansado"]:
            return "Aluno pode estar com rendimento baixo."
        else:
            return "Estado emocional indefinido."

    def exibir_dados(self):
        return f"{self.__nome} ({self.__idade} anos) - Humor: {self.__humor} | Situa√ß√£o: {self.verificar_situacao()}"


class Aluno_Especial_Neo(Humor_Aluno_Neo):
    def __init__(self, nome, idade, humor, atividade_extra):
        super().__init__(nome, idade, humor)
        self.__atividade_extra = atividade_extra

    def get_atividade_extra(self):
        return self.__atividade_extra

    # Polimorfismo: sobrescrevendo o m√©todo exibir_dados()
    def exibir_dados(self):
        return f"{self.get_nome()} ({self.get_idade()} anos) - Humor: {self.get_humor()} | Atividade: {self.__atividade_extra} | Situa√ß√£o: {self.verificar_situacao()}"


def inserir_aluno(nome, idade, humor, atividade_extra=None):
    cursor.execute("""
        INSERT INTO alunos (nome, idade, humor, atividade_extra)
        VALUES (%s, %s, %s, %s)
    """, (nome, idade, humor, atividade_extra))
    conexao.commit()
    print(f"‚úÖ Aluno '{nome}' inserido no banco com sucesso!")

def listar_alunos():
    cursor.execute("SELECT * FROM alunos")
    resultados = cursor.fetchall()
    print("üìã LISTA DE ALUNOS NO BANCO:")
    for aluno in resultados:
        print(aluno)

def excluir_aluno(id):
    cursor.execute("DELETE FROM alunos WHERE id = %s", (id,))
    conexao.commit()
    print(f"‚ùå Aluno com ID {id} exclu√≠do do banco.")



# Criando objetos
a1 = Humor_Aluno_Neo("Jo√£o", 16, "triste")
a2 = Aluno_Especial_Neo("Clara", 17, "feliz", "M√∫sica")

# Exibindo com polimorfismo e abstra√ß√£o
print(a1.exibir_dados())
print(a2.exibir_dados())

# Inserindo no banco
inserir_aluno(a1.get_nome(), a1.get_idade(), a1.get_humor())
inserir_aluno(a2.get_nome(), a2.get_idade(), a2.get_humor(), a2.get_atividade_extra())

# Listando
listar_alunos()

# Excluir aluno (exemplo: excluir o aluno de ID 1)
# excluir_aluno(1)

# Encerrar conex√£o
conexao.close()
```
